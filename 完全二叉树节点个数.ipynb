```python
#满二叉树，节点总数和树的高度呈指数关系：
public int countNode(TreeNode root){
    int h = 0;
    if(root == null) return 0;
    while(root != null){
        root = root.left;
        h++;
    }
    return (int)Math.pow(2,h) - 1;
}
```
